```python
import pandas as pd

df = pd.read_csv('messy_employee_data.csv')
print(df)

df.info()

df.describe()

df.tail()

df.head()

print(df.isnull())

print(df.isnull().sum())

df['Age'].fillna(df['Age'].mean(),inplace=True)

gp =df.groupby('Age')["Performance_Score"].mean()
print(gp)

df.dropna(subset=['Performance_Score'],inplace=True)

df.reset_index(drop=True,inplace=True)

df["Bonus"] = (df['Salary'] * .10) + df['Salary']

df["Performance_Grade"] = "None"

df.loc[df['Performance_Score'] >= 90, 'Performance_Grade'] = "A"
df.loc[(df['Performance_Score'] >= 75) & (df['Performance_Score'] <= 89), 'Performance_Grade'] = "B"
df.loc[df['Performance_Score'] < 75, 'Performance_Grade'] = "Needs Improvement"


grp = df.groupby("Department")

# grp_sal = df.groupby("Department")["Salary"].mean()

grp_sal = grp["Salary"].mean()
print(grp_sal)

grp_pscore = grp["Performance_Score"].mean()
print(grp_pscore)


df.sort_values(['Salary'],ascending=True)

import matplotlib.pyplot as plt

plt.bar(df["Years_At_Company"],df["Salary"])


plt.show()


plt.scatter(df["Years_At_Company"],df["Salary"])
plt.show()
```
