```python
import pandas as pd
```

```python
df = pd.read_csv("pokemon_data.csv")
print(df)
```

```python
print(df.head())
```

```python
# reading header
print(df.columns)
```

```python
# read each column
print(df["Speed"])

# for multiple  columns 
print(df[["Name","HP","Attack"]][0:10])
```

```python
# to read the row
print(df.iloc[1])

print(df.iloc[2:16])
```

```python
# Read specific location ---- Row , Column
df.iloc[3,2]
```

```python
# itter rows 
for index,row in df.iterrows():
    # print(index,row)
    print(index,row["Name"],row["Type 1"])
```

```python
# loc used for spedific conditions
print(df.loc[df['Type 1'] == "Water"])
```

```python
# sorting values
df.sort_values(['Type 1'],ascending=False)
print(df)
```

```python
# adding new column and summing update_display
df["total"] =  df["HP"] + df["Attack"]
print(df)
```

```python
df = df.drop(columns=['total'])
print(df)
```

```python
# to add new column and sum it up veritcally or horzontally - if vertical use axis = 0 and if horizontal use axis as 1
# summing up using iloc
df['total'] = df.iloc[:,4:7].sum(axis=1)
print(df)
# we can use reset index - 
```

```python
# TO INCLUDE NOT IN THE DF OPERATION WE CAN SIMPLY USE TILTE SYMBOL
# LIKE IF WE WANT TO FILTER THE TYPE WHICH IS NOT BUG THEN WE CAN USE IT 
print(df.loc[~df["Type 1"].str.contains("Bug")])
```

```python
# print(df.count(df['Type 1'] == 'Bug'))- incorrect
print(df['Type 1'].value_counts()["Bug"])  #will raise error if bug is not present
print((df["Type 1"] == "Bug").sum())
```

```python
# we can also use regex - regular expression 
import re 
df.loc[df['Type 1'].str.contains('fire|grass',flags=re.I,regex=True)] 
# what it does is that, it checks for fire or grass on  the type  1 and then ignore the alphabet casex and regex true means it applies the regular expression
```

```python
# we can give condition and change value if condition satisfies then change the value in both of the columns as different 
import re
df.loc[df["total"] > 200 & df["Type 1"].str.contains('grass',flags=re.I,regex=True),['Legendary']] = True
print(df.loc[df["Type 1"] == "Grass"])
```

```python
# Aggregate things

print(df.groupby(['Type 1'])[df.columns[4:10]].mean())
```

```python
grouped = df.groupby(['Type 1','Type 2']).sum()
print(grouped)
```
